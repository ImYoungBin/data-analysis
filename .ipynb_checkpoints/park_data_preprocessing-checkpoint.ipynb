```python
import pandas as pd
import geopandas as gpd
import numpy as np
import requests as rq
from urllib.parse import urlparse
from shapely.geometry import Polygon, LineString, Point
```

### 01. upload file and select columns

```python
date = '2021-12-23'
path = 'C:/Users/sys02/Downloads/전국도시공원정보표준데이터.csv'
df = pd.read_csv(path, encoding = 'euc-kr')
# crs = epsg:4326
df.info()
```

```python
df.columns
```

```python
# extract just needs
df = df[['관리번호', '공원명', '공원구분', '소재지도로명주소', '소재지지번주소', '경도', '위도', '공원면적']]
df.drop_duplicates(inplace = True)
df.reset_index(drop = True, inplace = True)
```

### 02. find miss coordinates

```python
miss = df[df['위도'].isna() | df['경도'].isna()]
miss.head()
```

```python
def naver_add_api(address):
    url = "https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?query="
    result = rq.get(urlparse(url+address).geturl(), headers={})
    json_obj = result.json()
    try:
        wgs = [json_obj['addresses'][0]['x'], json_obj['addresses'][0]['y']]
    except:
        wgs = [np.nan, np.nan]
    return wgs
```

```python
lst = []
for i in miss.index:
    if type(df.loc[i, '소재지도로명주소']) == float:
        wgs = naver_add_api(df.loc[i, '소재지지번주소'])
    else:
        wgs = naver_add_api(df.loc[i,'소재지도로명주소'])
    df.loc[i, '경도'] = wgs[0]
    df.loc[i, '위도'] = wgs[1]
```

```python
# 네이버 지도에서 검색
df.loc[14933, '소재지지번주소'] = '경기도 수원시 권선구 호매실동 1341번지'
df.loc[14739, '소재지지번주소'] = '전라북도 전주시 완산구 중노송동 253-75'
df.loc[12799, '소재지지번주소'] = '전라북도 익산시 함열읍 다송리 890'
df.loc[11775, '소재지지번주소'] = '경기도 광주시 중대동 239-1'
df.loc[10043, '소재지지번주소'] = '전라북도 전주시 완산구 서서학동 산20-1'
df.loc[8998, '소재지지번주소'] = '서울특별시 송파구 잠실동 319'
df.loc[8287, '소재지지번주소'] = '전라북도 남원시 주천면 송치리 1271-1'
df.loc[5651, '소재지지번주소'] = '충청남도 당진시 석문면 통정리 1960'
```

```python
lst = [5651, 8287, 8998, 10043, 11775, 12799, 14739, 14933]
for i in lst:
    wgs = naver_add_api(df.loc[i, '소재지지번주소'])
    df.loc[i, '경도'] = wgs[0]
    df.loc[i, '위도'] = wgs[1]
```

```python
df = df[df['경도'].notna()]
```

### 03. Transfer coordinates crs system and save

```python
df.loc[:, '경도'] = df.loc[:, '경도'].astype(float)
df.loc[:, '위도'] = df.loc[:, '위도'].astype(float)
geometry = [Point(xy) for xy in zip(df['경도'], df['위도'])]
df['geometry'] = geometry
# 좌표계 입력
crs = {'init':'epsg:4326'}
df = gpd.GeoDataFrame(df, crs = crs, geometry = 'geometry')
# 좌표계 변환
df = df.to_crs(epsg=5179)
```

```python
# column 이름을 영어로 바꾸어 줄 필요가 있어서 변환
df.rename(columns = {'관리번호' : 'manage_num', 
                     '공원명' : 'name', 
                     '공원구분' : 'type', 
                     '소재지도로명주소' : 'address_road', 
                     '소재지지번주소' : 'address_road', 
                     '경도' : 'x', 
                     '위도' : 'y', 
                     '공원면적' : 'area'}, inplace = True)
```

```python
# python version이 낮을 경우 protocol = 4로 저장된 pickle file만 열림
df.to_pickle('park_data_' + date + '.pkl.gz', compression = 'gzip', protocol = 4)
```
